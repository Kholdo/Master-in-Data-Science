# MAP en Python

```python
enteros = [0,1,2,3,4]
```

```python
def alcuadrado(num): return num ** 2
```

```python
cuadrados = map(alcuadrado, enteros)
cuadrados
```

```python
def suma_2_numeros(num_1, num_2): return num_1 + num_2

reduce(suma_2_numeros,cuadrados)
```

```python
#Ejercicio: implementar la funcion max como un reduce
#Es decir, encontrar el numero mas grande de una lista utilizando reduce().
lista = (1,32,576,56,4,3,2,3,1002)
```

```python
def max_function(num_1, num_2): 
    if num_1 >= num_2:
        return num_1
    else:
        return num_2
```

```python
maximo = reduce(max_function, lista)
maximo
```

```python
def h(a): return a % 2 == 0
filter(h, cuadrados)
```

```python
#funciones anónimas (functions literals)
```

```python
filter(lambda x : x % 2 == 0, cuadrados)
```

## Group by

```python
groups = aircraft_bien.groupby('manufacturer')

groups['nb_engines']

for k, group in groups['nb_engines']:
    print type(group)
```

```python
by_manufacturer = aircraft_bien.groupby('manufacturer')
engines_by_manufacturer = by_manufacturer['nb_engines']
avg_engines_per_manufacturer = engines_by_manufacturer.mean()
```

```python
engines_stats = engines_by_manufacturer.agg(['mean', 'std'])
```

```python
engines_stats
```

```python
engines_stats.sort_values(by=['mean', 'std'], ascending = False).head(10)
```

### JOINS

```python
aircraft_bien.join(avg_engines_per_manufacturer, rsuffix="_mean", on = 'manufacturer')
```
